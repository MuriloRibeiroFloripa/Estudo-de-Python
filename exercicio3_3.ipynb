```
# Crie uma função que desenhe um gráfico do tipo onda senoidal e outra que desenhe um gráfico
# do tipo onda quadrada. Deve ser informado o número de ciclos (repetições do comprimento de
# onda) e a amplitude.
import os

import ipywidgets as widgets
from ipywidgets import interact, interact_manual
import matplotlib.pyplot as plt
import numpy as np

def plot_data(x, y):
  plt.plot(x, y)
  plt.ylim(-6, 6)
  plt.show()

@interact(c = (1, 5), a = (1,5))
def plot_onda_semoidal(c, a):
  x = np.linspace(1, 2*c*np.pi,(20*c)+1)
  y = a * np.sin(x)
  plot_data(x, y)

@interact(c = (1,5), a = (1,5))
def plot_onda_quadrada(c, a):
  ref_x = np.array([0, 0, 1, 1, 1, 2, 2])
  x = np.empty(0)
  for i in range(c):
    x = np.append(x, ref_x+(i*2))
  y = np.tile(a * np.array([0, 1, 1, 0, -1, -1, 0]), c)
  plot_data(x,y)

  #############
  ciclos = 5
  amplitudes = 2


os.system("pause")



```
