```
# Crie um programa que gere dados no formato de uma parábola (ax^2 + bx + c) com ruído (+ d *
# np.random.randn(n)). Os dados devem ser gerados para o valor de 'n' dados de 10, 50 e 100
# pontos e plotados em diferentes plots (pyplot.plot).

import matplotlib.pyplot as plt
import numpy as np


num = [10, 50, 100] #numero de pontos, neste caso 50%, ele vai mostrar a data com esse valor

a= 1 # coeficiente = 1
b = 0 #coeficiente b =0
c = 0 # valor de baixo minimo do grafico, desloca eixo para cima

for n in num:
  d = n/2 #ruido

  ruido = d * np.random.randn(n+1)

  x = np.arange(-n/2, (n/2)+1)  #np.arange faz uma divisao dos pontos, faz a distribuição
  # print(x,"(101 pontos distribuidos, definindo a sequencia de numeros)")
  y = a*(x**2) + b*x + c + ruido #y gera 
  # print()
  # print(y)
  # print()
  plt.plot(x,y, label="data=" + str(n)) # cria o plot com os valores do eixo x e y, com label data.
  plt.xlabel('eixo x') # mostra os calores eixo x
  plt.ylabel('eixo y') # mostra os calores eixo y
  plt.legend() # mostra legenda com base no label que colocamos
  plt.show() # mostra









```
